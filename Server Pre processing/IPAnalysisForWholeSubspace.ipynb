```python
import pandas as pd
```

```python
for i in range(7):
    c = pd.read_csv('FinalServerIPCount_Instance_no_0'+str(i+1)+'_round_3.csv')
    c.to_csv('FinalServerIPCountNotGrouped.csv', mode='a', header=False,index=False)
```

```python
df = pd.read_csv('FinalServerIPCountNotGrouped.csv',names = ["Servers", "IPCount"])
```

```python
FinalServerIPCount = df.groupby(['Servers']).sum().reset_index()
```

```python
FinalServerIPCount = FinalServerIPCount.sort_values(by=[ 'IPCount'],ascending=False)
```

```python
FinalServerIPCount.to_csv('FinalServerIPCountGrouped_round_3.csv',index=False)
```

```python
FinalServerIPCount.iloc[:10,[1]].values.tolist()[5][0]
```

```python
listsize = 20
```

```python
import matplotlib.pyplot as plt

floatipcount = []
for i in range(listsize):
  floatipcount.append(FinalServerIPCount.iloc[:listsize,[1]].values.tolist()[i][0])

ListOfServers = [i[0] for i in FinalServerIPCount.iloc[:listsize,[0]].values.tolist()]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
vendors = FinalServerIPCount.iloc[:listsize,[0]].values.tolist()
ipcount = FinalServerIPCount.iloc[:listsize,[1]].values.tolist()
ax.barh(ListOfServers,floatipcount,height=0.7)
ax.set_xlabel('IP Count')
ax.set_ylabel('Servers')
ax.set_title('Count Of IPs running various servers ')
y_pos = range(len(vendors))
plt.show()
```
