```python
def getWhoIsDataPerIp(ip,result):
    keys =  ["ip","No Of Owners", "CIDR", "netType", "Organization","Most Recent Update","No of Who is Records"]
    details = dict.fromkeys(keys)

    details['ip'] = ip
    details['No of Who is Records'] = result.get('meta') and result.get('meta').get('count')
    ownerCount = 0

    try:
          if(result.get('data') is not None and result.get('meta').get('count')>0):

            for i in range(result.get('meta').get('count')):
              if(result.get('data')[i].get('attributes').get('last_updated')):
                details['Most Recent Update'] = result.get('data')[i].get('attributes').get('last_updated')
                break

            for i in range(result.get('meta').get('count')):
              if(result.get('data')[i].get('attributes').get('whois_map').get('NetType')):
                details['netType'] = result.get('data')[i].get('attributes').get('whois_map').get('NetType')
                break

            orglist = ['OrgName','Organization','Admin Organization','Tech Organization']

            for j in range(result.get('meta').get('count')):
              for i in range(4):
                if(result.get('data')[j].get('attributes').get('whois_map').get(orglist[i]) is not None):
                  details['Organization'] = result.get('data')[j].get('attributes').get('whois_map').get(orglist[i])
                  break

            for i in range(result.get('meta').get('count')):
              if(result.get('data')[i].get('attributes').get('whois_map').get('CIDR')):
                details['CIDR'] = result.get('data')[i].get('attributes').get('whois_map').get('CIDR')
                break

            for i in range(result.get('meta').get('count')):
              if(result.get('data')[i].get('attributes').get('whois_map').get('OrgName')):
                ownerCount=ownerCount+1

            details['No Of Owners'] = ownerCount
    except:
        print('Error Occured')
        return details
    finally:
        return details
```
