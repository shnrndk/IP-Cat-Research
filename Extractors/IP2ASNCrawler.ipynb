```python
df = pd.read_csv('vpnasnsummary.csv')
```

```python
df = df.sample(n=10000, random_state=1)
```

```python
df['ip'].values
```

```python
ipasn = []
for ip in df['ip'].values:
    r =requests.get(f'http://127.0.0.1/v1/as/ip/{ip}')
    result = json.loads(r.text)
    if 'as_description' in result:
        asname = json.loads(r.text)['as_description']
    else:
        asname = 'null'
    if 'as_number' in result:
        asnum = json.loads(r.text)['as_number']
    else:
        asnum = 'null'
    ipasn.append([ip,asname,asnum])
```

```python
ipasn
```

```python
dfnew = pd.DataFrame(ipasn)
dfnew.columns = ['ip','as_desc','as_no']
```
