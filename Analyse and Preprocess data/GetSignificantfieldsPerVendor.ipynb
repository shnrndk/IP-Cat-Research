```python
batch_no=1

for chunk in pd.read_csv('Data/data.csv',chunksize=100000):
  chunk.to_csv('Data/100KChunks/Banners'+str(batch_no)+'.csv',index=False)
  batch_no+=1
    
no_of_chunks = batch_no-1
```

```python
no_of_chunks
```

```python
vendoriplist = pd.read_csv('hosting_expanded.csv',names=['ip'])
vendoriplist
```

```python
for i in range(no_of_chunks):
    df = pd.read_csv('Data/100KChunks/Banners'+str(i+1)+'.csv')
    merged = pd.merge(df, vendoriplist)
    fieldcountdf = pd.DataFrame(merged.count().reset_index())
    fieldcountdf.columns = ['Fields','count']
    fieldcountdf = fieldcountdf.set_index('Fields')
    df = fieldcountdf.sort_values(by='count', ascending=False)
    df.to_csv('Data/100KChunks/Temp'+str(i+1)+'.csv')
```

```python
df1 = pd.read_csv('Data/100KChunks/Temp'+str(i+1)+'.csv')
df1 = df1.set_index('Fields')
for i in range(no_of_chunks):
    df = pd.read_csv('Data/100KChunks/Temp'+str(i+2)+'.csv')
    df = df.set_index('Fields')
    df1 = df1.add(df, fill_value=0)
    if(i+2==no_of_chunks):
        break
df1 = df1.reset_index()
df1 = df1.sort_values(by='count', ascending=False)
```

```python
df1.to_csv('FieldCount_'+instance_no+'.csv',index=False)
```
