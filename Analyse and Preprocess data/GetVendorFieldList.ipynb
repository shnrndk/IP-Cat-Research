```python
j=0;
for i in range(no_of_chunks):
    df = pd.read_csv('Data/100KChunks/GCPmerged'+str(i+1)+'.csv')
    merged = pd.merge(df, vendoriplist)
    temp =  merged.filter(fields)
    if(temp.shape[0]>0):
        temp.to_csv('Data/100KChunks/Temp'+str(j+1)+'.csv')
        j=j+1
```

```python
temp = []
for i in range(no_of_chunks):
    df = pd.read_csv('Data/100KChunks/Temp'+str(i+1)+'.csv')
    temp.append(df)
GCPData=pd.concat(temp, ignore_index=True)
```

```python
GCPData = GCPData.drop(columns=['Unnamed: 0'])
GCPData.to_csv(service+'FinalData'+instance_no+'.csv',index=False)
```
