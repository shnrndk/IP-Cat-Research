```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
batch_no=1

for chunk in pd.read_csv('Data/data.csv',chunksize=100000):
  chunk.to_csv('Data/100KChunks/Banner'+str(batch_no)+'.csv',index=False)
  batch_no+=1
    
no_of_chunks = batch_no-1
```

```python
# no_of_chunks = 20
# instance_no = '07'
round_no = '3'
```

```python
for i in range(no_of_chunks):
    print(i)
    c = pd.read_csv('Data/100KChunks/Banner'+str(i+1)+'.csv')
    c = c.dropna(subset=['data.http.result.response.headers.server.0'],axis = 0)
    c = c[['ip','data.http.result.response.headers.server.0']]
    c.columns = ['IP','Server']
    c = c.sort_values(by=[ 'Server'])
    c.to_csv('Data/100KChunks/TempIPlistPerServer'+str(i+1)+'.csv',index=False)
```

```python
for i in range(no_of_chunks):
    c = pd.read_csv('Data/100KChunks/TempIPlistPerServer'+str(i+1)+'.csv')
    c.to_csv('Data/100KChunks/TempIPlistPerServer.csv', mode='a', header=False,index=False)
```

```python
df = pd.read_csv('Data/100KChunks/TempIPlistPerServer.csv',names = ["IP", "Server"])
df = df.sort_values(by=[ 'Server'])
```
